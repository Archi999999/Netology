```python
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# Задание №1

KEYWORDS = ['python', 'парсинг']

def news(KEYWORDS):
    URL = 'https://habr.com/ru/search/'
    news = pd.DataFrame()
    for KEYS in KEYWORDS: 
        params = {'q': KEYS}
        res = requests.get(URL, params)
        soup = BeautifulSoup(res.text)
        news_blocks = soup.find_all('article', class_="tm-articles-list__item")
        for news_ in news_blocks:
            date = news_.find('time').get('title')
            link = news_.find('a').get('href')
            link = 'https://habr.com' + link
            title = str(news_.find('a', class_='tm-article-snippet__title-link'))     
            title = re.sub(r'<[^>]+>','',title)
            row = {'date': date, 'title': title, 'link': link}
            news = pd.concat([news, pd.DataFrame([row])], ignore_index=True) 
    return news
news(KEYWORDS)
```
