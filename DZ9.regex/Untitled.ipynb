```python
#ЗАДАНИЕ 1

import re

def auto_numbers():
    car_id = input('car_id=')
    car_number = r'[укенхваролсмт]\d{3}[укенхваролсмт]{2}\d{2,3}'
    if re.match(car_number, car_id, flags=re.IGNORECASE):
        number = re.findall(r'\w{6}', car_id)
        region = re.findall(r'(?<=[а-хА-Х]{2})\d{2,3}', car_id)
        print(f'Результат: {number}  валиден. Регион: {region}')
    else:
        print('Результат: Номер не валиден')
        
auto_numbers()


# Вариант лучше
import re

car_id = input('Введите номер: ')

def check(carid):
    pattern = r'([АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2})(\d{2,3})'
    result = re.match(pattern, carid)
    if result:
        print(f'Номер {result.group(1)} валиден. Регион: {result.group(2)}')
    else:
        print('Номер не валиден')

check(car_id)
```

```python
#ЗАДАНИЕ 4
import re

emails = 'test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru, xyz@gmail.com'

def domens(emails):
    dict_doms = {}
    for i in re.findall(r'@(\w*\.\w*)', emails): 
        if i not in dict_doms:
            dict_doms[i] = 1
        else:
            dict_doms[i] += dict_doms[i]
    return dict_doms

        
domens(emails)    

```
