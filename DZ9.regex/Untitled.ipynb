```python
#ЗАДАНИЕ 1

import re

def auto_numbers():
    car_id = input('car_id=')
    car_number = r'[укенхваролсмт]\d{3}[укенхваролсмт]{2}\d{2,3}'
    if re.match(car_number, car_id, flags=re.IGNORECASE):
        number = re.findall(r'\w{6}', car_id)
        region = re.findall(r'(?<=[а-хА-Х]{2})\d{2,3}', car_id)
        print(f'Результат: {number}  валиден. Регион: {region}')
    else:
        print('Результат: Номер не валиден')
        
auto_numbers()
```

```python
#ЗАДАНИЕ 2

some_string = '''Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'''

def remove_repeats(text):
    return re.sub(r'(\s\b\w+)\1+', r'\1', text)
    
print(remove_repeats(some_string))
```
