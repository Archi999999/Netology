```python
#Задание 2


queries = [
'смотреть сериалы онлайн',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про спорт',
]

lines = (len(queries))
words_2 = 0
words_3 = 0
for i in range(lines):
    if (len(queries[i].split())) == 2:
        words_2 +=1
    elif (len(queries[i].split())) == 3:
        words_3 +=1
print ('Поисковых запросов, содержащих 2 слов(а): ', round(100/lines*words_2,2),'%')
print ('Поисковых запросов, содержащих 3 слов(а): ', round(100/lines*words_3,2),'%')
```
