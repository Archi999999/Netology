```python
import json
purchases ={}
with open('C:/Python/Netology/x-files/downloads/purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in f:
        line = json.loads(line)
        print (line)
        purchases[line['user_id']] = line['category']
```

```python
x=0
with open('C:/Python/Netology/x-files/downloads/visit_log.csv', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip().split(',')
        for key, value in purchases.items():
            if line[0] in key:
                g = open('C:/Python/Netology/x-files/downloads/funnel.csv', 'w', encoding='utf-8')
                g.write(line)
                g.write(value)
                g.close()
        x+=1
        if x >25:
            break
            
            
```
