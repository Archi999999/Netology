```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}


def commands():
    print('''Список команд:
    “p” может узнать владельца документа по его номеру
    “s” может по номеру документа узнать на какой полке он хранится
    “l” может увидеть полную информацию по всем документам
    “ads” может добавить новую полку
    “ds” может удалить существующую полку из данных (только если она пустая)
    “q” для выхода из программы
    ''')

def command_p(docs):
    command = input('Введите номер документа: ')
    for dicts in docs:
        if command in dicts.values():
            print('Владелец документа: ', dicts['name'])
            return command
    print('Документ не найден в базе')
    return command

def command_s(docs, dirs):
    command = input('Введите номер документа: ')
    for key, value in dirs.items():
        if command in value:
            print('Документ хранится на полке: ', key)
            return command
    print('Документ не найден в базе')
    return command

def command_l(docs, dirs):
    for dicts in docs:
        for key, value in dirs.items():
            if dicts['number'] in value:
                print('№: ',dicts['number'], ', тип: ', dicts['type'], ', владелец: ', dicts['name'], ', полка хранения: ', key)
    return

def command_ads(docs, dirs):
    command = input('Введите номер полки: ')    
    for key, value in dirs.items():
        if command == key:
            print('Такая полка уже существует. Текущий перечень полок:', list(dirs.keys()))
            return command
        else:
            dirs[command] = []
            print('Полка добавлена. Текущий перечень полок: ', list(dirs.keys()))
            return command
        
def command_ds(docs, dirs):
    command = input('Введите номер полки: ')
    for key, value in dirs.items():
        if command == key and value == []:
            del(dirs[key])
            print('Полка удалена. Текущий перечень полок: ', list(dirs.keys()))
            return command
        elif command == key:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', list(dirs.keys()))
            return command
    print('Такой полки не существует. Текущий перечень полок: ' ,list(dirs.keys()))
    return command
    

def main(docs = None, dirs = None):
    while True:
        command = input('Введите команду: ')
        if command not in ['p','s','l','ads','ds','q']:
            print('КОМАНДА НЕ ВЕРНА!')
            commands()
        elif command == 'p':
            command_p(docs)
        elif command == 's':
            command_s(0, dirs)
        elif command == 'l':
            command_l(docs, dirs)
        elif command == 'ads':
            command_ads(0, dirs)
        elif command == 'ds':
            command_ds(0, dirs)
        elif command == 'q':
            print('программа завершена!')
            break

            
            
            
commands()            
main(documents, directories)          
```
