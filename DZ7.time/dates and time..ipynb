```python
from datetime import datetime

The_Moscow_Times = 'Wednesday, October 2, 2002'
The_Guardian = 'Friday, 11.10.13'
Daily_News = 'Thursday, 18 August 1977'

The_Moscow_Times = datetime.strptime(The_Moscow_Times, '%A, %B %d, %Y') 
The_Guardian = datetime.strptime(The_Guardian, '%A, %d.%m.%y')
Daily_News = datetime.strptime(Daily_News, '%A, %d %B %Y')

print(The_Moscow_Times)
print(The_Guardian)
print(Daily_News)
```

```python
stream = ['2018-04-02', '2018-02-29', '2018-19-02']
date_formats = ['%Y-%m-%d', '%Y-%d-%m']
def check(list_):
    for format_ in date_formats:
        for dates in list_:       
            try:
                dates = datetime.strptime(dates, format_)
                print(dates, 'True')
            except:
                print(dates, 'False')
        return
        
check(stream)

```

### ВОПРОС:
###     Не удается произвести проверку третьей даты ('2018-19-02') по второму формату дат ('%Y-%d-%m').

```python
from datetime import timedelta
from datetime import datetime

list_date =[]
def date_range(start_date, end_date):
    try:
        start_date = datetime.strptime(start_date, '%Y-%m-%d')
        end_date = datetime.strptime(end_date, '%Y-%m-%d')
        while start_date <= end_date:
            list_date.append(start_date.strftime('%Y-%m-%d'))        
            start_date += timedelta(days=1)
        return
    except:
        return
                
start_date = '2018-01-01'
end_date = '2018-01-07'        
date_range(start_date, end_date)        
   
print(list_date)
```
