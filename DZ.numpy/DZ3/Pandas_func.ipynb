```python
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class
```

```python
# Задание 1

ratings = pd.read_csv('C:/Python/Netology/x-files/ml-latest-small/ratings.csv')

def classific(rating):
    if int(rating) <= 2:
        ratings['class'] = 'низкий рейтинг'
    elif int(rating) <= 4:
        ratings['class'] = 'средний рейтинг'
    elif int(rating) <= 5:
        ratings['class'] = 'высокий рейтинг'
    return   
        
ratings['rating'].head().apply(classific)
ratings
```

```python
# Задание 2

keywords = pd.read_csv('C:/Python/Netology/x-files/ml-latest-small/keywords.csv')
keywords.head()

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}


def regions(inquiry):
    '''гео-классификатор, который каждой строке сможет выставить географическую принадлежность
    определенному региону. Т. е. если поисковый запрос содержит название города региона,
    то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит
    названия города, то ставим ‘undefined’.'''
    
    
    for reg, cities in geo_data.items():
        for city in cities:
            if city in inquiry:
                return reg
            else:
                return 'undefined'
    
keywords['region'] = keywords['keyword'].apply(regions)

keywords.sort_values(by = 'region').tail()
```
