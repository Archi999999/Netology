```python
#Задание № 1

logs = pd.read_csv('C:/Python/Netology/x-files/visit_log.csv', sep = ';')
logs.head()

def source_type(df):
    if df['traffic_source'] == ('yandex' or 'google') :
        return 'organic'
    elif df['region'] == 'Russia' and df['traffic_source'] == 'paid':
        return 'ad'
    elif df['region'] != 'Russia' and df['traffic_source'] == 'paid':
        return 'other'
    else:
        return df['traffic_source']   

logs['source_type'] = logs.apply(source_type, axis = 1)

logs.tail()
```

```python
# Задание № 3

ratings = pd.read_csv('C:/Python/Netology/x-files/ml-latest-small/ratings.csv')

#'''считаем кол-во оценок для каждого пользователя'''
filt_100 = ratings.groupby(['userId']).count()[['rating']].reset_index()

#'''оставляем только со 100 и выше оценок'''
filt_100 = filt_100[filt_100['rating'] >= 100]

#'''убираем лишние столбцы'''
filt_100 = filt_100[['userId']]

#'''соединяем с основной таблицей, чтобы остались только нужные пользователи'''
filt_100 = filt_100.merge(ratings, how = 'left')


filt_100.groupby(['userId', 'timestamp']).mean().head()


# filt_100.head()
```
