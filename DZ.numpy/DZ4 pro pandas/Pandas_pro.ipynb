```python
#Задание № 1

logs = pd.read_csv('C:/Python/Netology/x-files/visit_log.csv', sep = ';')
logs.head()

def source_type(df):
    if df['traffic_source'] == ('yandex' or 'google') :
        return 'organic'
    elif df['region'] == 'Russia' and df['traffic_source'] == 'paid':
        return 'ad'
    elif df['region'] != 'Russia' and df['traffic_source'] == 'paid':
        return 'other'
    else:
        return df['traffic_source']   

logs['source_type'] = logs.apply(source_type, axis = 1)

logs.tail()

# В первом задании можно также использовать вариант решения через метод loc. Например, для первого правила:
# logs.loc[logs.traffic_source.isin(['yandex', 'google']), 'traffic_type'] = 'organic'
```

```python
# Задание № 3

ratings = pd.read_csv('C:/Python/Netology/x-files/ml-latest-small/ratings.csv')

ratings = ratings.groupby('userId').agg({'timestamp': ['min', 'max', 'count']})
ratings = ratings[ratings['timestamp']['count']>= 100]
ratings['live_time'] = ratings['timestamp'].apply(lambda x: x['max'] - x['min'],axis = 1)
ratings.head()
```

```python
import datetime
ratings['live_time'] = ratings['live_time'].apply(datetime.timedelta(seconds=ratings['live_time']))
```
