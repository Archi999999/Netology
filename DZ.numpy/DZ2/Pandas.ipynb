```python
#'''поиск названия стран в датафрейме'''

power = pd.read_csv('C:/Python/Netology/x-files/Pandas/power.csv')
power[ power['country'].str.contains('lit', case=False) ]['country'].unique()
#'Latvia'
#'Lithuania'
#'Estonia'
```

```python
filtered_countries = power[ (power['country']=='Latvia') | (power['country']=='Lithuania') | (power['country']=='Estonia') ]
filtered_quantity = filtered_countries[(filtered_countries['quantity'] > 0) & (filtered_countries['year'] > 2005) & (filtered_countries['year'] < 2014)]
filtered_categories = filtered_quantity[(filtered_quantity['category'] == 4) | (filtered_quantity['category'] == 12) | (filtered_quantity['category'] == 21)]
filtered_categories['quantity'].sum()
```
